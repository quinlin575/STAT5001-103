```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def problem1():
    data = pd.read_csv("crime_data.csv", index_col="Year", header=0)
    datapercapita = data.drop(columns=["Population"]).div(.00001 * data["Population"], axis=0)
    data.plot(kind="hist", y=["Total"], bins=10, title="Frequency of Amount of Total Crimes")
    datapercapita.drop(columns=["Total", "Property", "Burglary", "Larceny", "Vehicle Theft", "Violent"]).plot(title="Violent Crimes(Per 100,000) Over Time")
    datapercapita.drop(columns=["Total", "Property", "Murder", "Forcible Rape", "Robbery", "Aggravated Assault", "Violent"]).plot(title="Property Crimes(Per 100,000) Over Time")
```

```python
def problem2():
    data = pd.read_csv("crime_data.csv", index_col="Year", header=0)
    data.plot(kind="scatter", x="Larceny", y="Violent", title="Larceny vs Violent Crimes")
    data.plot(kind="scatter", x="Larceny", y="Burglary", title="Larceny vs Burglary")
    data.plot(kind="scatter", x="Larceny", y="Aggravated Assault", title="Larceny vs Aggravated Assault")
    return "Burglary"
```

```python
def problem3():
    data = pd.read_csv("crime_data.csv", index_col="Year", header=0)
    data.plot(kind="box", y=["Burglary", "Violent", "Vehicle Theft"], title="Distributions for Different Crimes")
    data.plot(kind="hexbin", x="Vehicle Theft", y="Robbery", gridsize=10, title="Vehicle Theft vs Robbery Hexbin")
```

```python
def problem4():
    data = pd.read_csv("college.csv", header=0)
    point_colors = ['red' if point == "Yes" else 'blue' for point in data['Private']]
    data.plot(kind="scatter", x="S.F.Ratio", y="Outstate", s=3, color=point_colors, title="Student Facult Ratio vs Tuition\nPrivate: Red, Public: Blue")
    plt.figtext(0, -.05,  "James, G., Witten, D., Hastie, T., and Tibshirani, R. (2013) An Introduction to Statistical\nLearning with applications in R, www.StatLearning.com, Springer-Verlag, NewYork", fontsize=7)
    data.plot(kind='hist', y=["PhD"], bins=range(0,101,10), title="Frequency of Percent of Faculty with PhD")
    data.plot(kind="scatter", x="Outstate", y="Expend", title="Tuition vs Per Student Expenditure", s=3, color="black")
```
