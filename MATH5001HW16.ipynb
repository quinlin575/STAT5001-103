```python
import numpy as np
import matplotlib.pyplot as plt
import scipy


mlist = [5, 25, 125]

x = np.linspace(0, 10, 1000)
true_cdf = scipy.stats.expon.cdf(x, scale=1)

plt.figure(figsize=(10, 6))
plt.plot(x, true_cdf, label=r'True CDF $F_X(x)$', color='black', linestyle='--')
for m in mlist:
    u_values = np.arange(1, m + 1) / m
    x_m_samples = -np.log(1 - u_values)

    x_m_samples.sort()
    empirical_cdf = np.arange(1, m + 1) / m
    
    plt.step(x_m_samples, empirical_cdf, where='post', label=f'm = {m}')

plt.title('Convergence of Empirical CDF to True CDF')
plt.xlabel('x')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()

```
