```python
import numpy as np
import random
from matplotlib import pyplot as plt

nlist = [5, 25, 125, 625]
plt.figure(figsize=(10,6))


for i in nlist:
    rolls = []
    rollcount = np.array([0,0,0,0,0,0])
    cdfrolls = np.array([0,0,0,0,0,0])
    xvals = np.array([1,2,3,4,5,6])
    for j in range(i):
        roll = random.randint(1,6)
        rolls.append(roll)
    rolls.sort()
    print(rolls)
    for j in range(6):
        runningtotal = 0
        for k in rolls:
            if k == j+1:
                runningtotal += 1
        rollcount[j] = runningtotal
    for index, j in enumerate(rollcount):
        cdfrolls[index] = np.sum(rollcount[:index + 1])
    print(max(np.max(rollcount / i - 1 / 6), np.max(1 / 6 - rollcount / i)))
    plt.step(xvals, cdfrolls / i, where='post', label=f'n = {i}')


plt.title('1-6 Dice Roll CDF')
plt.xlabel('x')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()


```
