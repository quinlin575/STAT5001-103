```python
def problem2(arr):
    squeezed = []
    for i in arr:
        squeezed.append(np.squeeze(i))
    
    maxr = 0
    maxc = 0
    for i in squeezed:
        r = np.shape(i)[0]
        c = np.shape(i)[1]
        if r > maxr:
            maxr = r
        if c > maxc:
            maxc = c
    
    paddedr = []
    for i in squeezed:
        r = np.shape(i)[0]
        c = np.shape(i)[1]
        if r < maxr:
            paddedr.append(np.vstack((i, np.zeros((maxr - r, c)))))
        else:
            paddedr.append(i)
    paddedc = []
    for i in paddedr:
        r = np.shape(i)[0]
        c = np.shape(i)[1]
        if c < maxc:
            paddedc.append(np.hstack((i, np.zeros((r, maxc - c)))))
        else:
            paddedc.append(i)
    final = np.dstack(paddedc)
    return final
```
