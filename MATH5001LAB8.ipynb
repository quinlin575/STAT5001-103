```python
import numpy as np
import sympy as sy
from matplotlib import pyplot as plt
```

```python
def problem1():
    x = sy.symbols('x')
    y = sy.symbols('y')
    return sy.Rational(2,5) * sy.exp(x**2 - y) * sy.cosh(x + y) + sy.Rational(3,7) * sy.log(x * y + 1)
```

```python
def problem2():
    i, j, x, y = sy.symbols('i j x y')
    return sy.simplify(sy.product(sy.summation(j * (sy.sin(x) + sy.cos(x)), (j, i, 5)), (i, 1, 5)))
```

```python
def problem3(N):
    x, n, y = sy.symbols('x n y')
    expr = sy.summation(x ** n / sy.factorial(n), (n, 0, N))
    newexpr = expr.subs(x, -y**2)
    f = sy.lambdify(y, newexpr, "numpy")
    domain = np.linspace(-2,2,101)
    plt.plot(domain, f(domain), label="Maclaurin Series", alpha=.5, color='red')
    plt.plot(domain, np.exp(-1 * domain **2), label="True Value", alpha=.5, color='blue')
    plt.legend()
```

```python
def problem4():
    x, y, r, theta = sy.symbols('x y r theta')
    expr = ((x**2 + y**2)**sy.Rational(7,2) + 18 * x**5 * y - 60 * x**3 * y**3 + 18 * x * y**5) / (x**2 + y**2)**3
    newexpr = expr.subs({x: r * sy.cos(theta), y: r * sy.sin(theta)})
    newexpr = sy.simplify(newexpr)
    answer = sy.solve(newexpr, r)[0]
    f = sy.lambdify(theta, answer, "numpy")
    domain = np.linspace(0, 2*np.pi, 500)
    plt.plot(f(domain) * np.cos(domain), f(domain) * np.sin(domain))

```

```python
def problem5():
    x, y, l = sy.symbols('x y l')
    M = sy.Matrix([[x-y, x, 0], [x, x-y, x], [0, x, x-y]])
    evals = sy.solve(sy.Eq((M - l*sy.eye(3)).det(), 0), l)
    edict = {}
    for e in evals:
        ematrix = M - e * sy.eye(3)
        null = ematrix.nullspace()
        edict[e] = null
    return edict
    
```

```python
def problem6():
    x = sy.symbols('x')
    expr = 2*x**6 - 51*x**4 + 48*x**3 + 312*x**2 - 576*x - 100
    f = sy.lambdify(x, expr, "numpy")
    df = sy.diff(expr, x)
    ddf = sy.diff(expr, x, x)
    crits = sy.solve(df, x)
    mins = []
    maxs = []
    for i in crits:
        if ddf.evalf(subs={x: i}) > 0:
            mins.append(i)
        elif ddf.evalf(subs={x: i}) < 0:
            maxs.append(i)
    domain = np.linspace(-5, 5, 1001)
    plt.plot(domain, f(domain))
    for i in mins:
        plt.plot(i, f(i), marker='o', color="red", label="Minimum")
    for i in maxs:
        plt.plot(i, f(i), marker='o', color="black", label="Maximum")
    plt.legend()
    return mins, maxs
```

```python
def problem7():
    r, rho, theta, phi = sy.symbols('r rho theta phi')
    x = rho * sy.sin(phi) * sy.cos(theta)
    y = rho * sy.sin(phi) * sy.sin(theta)
    z = rho * sy.cos(phi)

    f = (x**2 + y**2 + z**2) **2
    J = sy.Matrix([x, y, z]).jacobian([rho, phi, theta])
    detJ = J.det()

    integrand = sy.simplify(f * detJ)
    volume = sy.integrate(integrand, (rho, 0, r), (phi, 0, sy.pi), (theta, 0, 2 * sy.pi))
    print(volume)

    v = sy.lambdify(r, volume, "numpy")
    domain = np.linspace(0,3,301)
    plt.plot(domain, v(domain))
    return v(2)


```
